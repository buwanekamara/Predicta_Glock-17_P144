```python
import pandas as pd
from prophet import Prophet
import numpy as np

# Load data
df = pd.read_csv('/content/historical_weather.csv')

# Group the DataFrame by 'city_id'
grouped = df.groupby('city_id')

# Create a dictionary to hold DataFrames for each city
city_datasets = {city_id: group_df for city_id, group_df in grouped}

# Define a function to fill NaN values using rolling mean
def fill_na_with_rolling_mean(series, window=7):
    return series.fillna(series.rolling(window, min_periods=1).mean())

# Prepare the submission DataFrame
submission_df = pd.DataFrame(columns=['submission_ID', 'city_id', 'date', 'avg_temp_c'])
submission_id = 1

for city_id, df in city_datasets.items():
    print(f"Processing city: {city_id}")

    # Convert date column to datetime
    df['date'] = pd.to_datetime(df['date'])
    # Set date as index
    df.set_index('date', inplace=True)

    # Apply the function to each relevant column
    df['avg_temp_c'] = fill_na_with_rolling_mean(df['avg_temp_c'])
    df['min_temp_c'] = fill_na_with_rolling_mean(df['min_temp_c'])
    df['max_temp_c'] = fill_na_with_rolling_mean(df['max_temp_c'])
    df['avg_wind_dir_deg'] = fill_na_with_rolling_mean(df['avg_wind_dir_deg'])

    # Transform wind direction to sine and cosine components
    df['wind_dir_sin'] = np.sin(np.deg2rad(df['avg_wind_dir_deg']))
    df['wind_dir_cos'] = np.cos(np.deg2rad(df['avg_wind_dir_deg']))

    # Ensure no NaN values remain
    df.fillna(0, inplace=True)

    # Reset the index to get back the date column
    df.reset_index(inplace=True)

    # Initialize the new DataFrame for predictions
    future_dates = pd.date_range(start='2019-01-01', periods=7, freq='D')
    new_df = pd.DataFrame({'date': future_dates, 'city_id': city_id})

    # Columns to forecast
    columns_to_forecast = ['avg_temp_c']

    # Train and predict for each column
    for column in columns_to_forecast:
        print(f"Training model for {column} for city {city_id}")
        temp_df = df[['date', column, 'min_temp_c', 'max_temp_c', 'wind_dir_sin', 'wind_dir_cos']].rename(columns={'date': 'ds', column: 'y'})

        # Train the model
        m = Prophet()
        m.add_regressor('min_temp_c')
        m.add_regressor('max_temp_c')
        m.add_regressor('wind_dir_sin')
        m.add_regressor('wind_dir_cos')

        m.fit(temp_df)

        # Create future dataframe with additional regressors
        future = m.make_future_dataframe(periods=7)
        future = future.merge(df[['date', 'min_temp_c', 'max_temp_c', 'wind_dir_sin', 'wind_dir_cos']], left_on='ds', right_on='date', how='left')
        future.drop(columns=['date'], inplace=True)

        # Fill NaN values in future dataframe
        future['min_temp_c'].fillna(future['min_temp_c'].mean(), inplace=True)
        future['max_temp_c'].fillna(future['max_temp_c'].mean(), inplace=True)
        future['wind_dir_sin'].fillna(future['wind_dir_sin'].mean(), inplace=True)
        future['wind_dir_cos'].fillna(future['wind_dir_cos'].mean(), inplace=True)

        # Make future predictions
        forecast = m.predict(future)

        # Extract the predictions for the next 7 days
        forecast = forecast[['ds', 'yhat']].tail(7).reset_index(drop=True)
        new_df[column] = forecast['yhat']

    # Combine the original and new DataFrames
    df['city_id'] = city_id
    df = df.rename(columns={'date': 'ds'})

    # Reorder columns to match the new DataFrame format
    df = df[['ds', 'city_id', 'avg_temp_c', 'min_temp_c', 'max_temp_c', 'wind_dir_sin', 'wind_dir_cos']]

    # Concatenate the original and new DataFrames
    final_df = pd.concat([df, new_df.rename(columns={'date': 'ds'})], ignore_index=True)

    # Ensure no NaN values remain
    final_df.fillna(0, inplace=True)
    # Reset the index to get back the date column
    final_df.reset_index(inplace=True)

    final_df = final_df[['ds', 'avg_temp_c', 'min_temp_c', 'max_temp_c', 'wind_dir_sin', 'wind_dir_cos']]
    final_df.columns = ['ds', 'y', 'x1', 'x2', 'x3', 'x4']

    train = final_df.iloc[:len(final_df) - 7]
    test = final_df.iloc[len(final_df) - 7:]

    m = Prophet()

    m.add_regressor('x1')
    m.add_regressor('x2')
    m.add_regressor('x3')
    m.add_regressor('x4')

    m.fit(train)
    future = m.make_future_dataframe(periods=7)
    future['x1'] = final_df['x1']
    future['x2'] = final_df['x2']
    future['x3'] = final_df['x3']
    future['x4'] = final_df['x4']

    forecast = m.predict(future)
    forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(7)

    # Add predicted values to submission DataFrame
    submission_data = {
        'submission_ID': range(submission_id, submission_id + 7),
        'city_id': city_id,
        'date': future_dates,
        'avg_temp_c': new_df['avg_temp_c']
    }
    submission_id += 7
    submission_df = pd.concat([submission_df, pd.DataFrame(submission_data)], ignore_index=True)

# Save to CSV
submission_df.to_csv('submission2.csv', index=False)

print("Forecasting and submission file generation complete.")
```

```python
import pandas as pd
from prophet import Prophet
from statsmodels.tsa.seasonal import seasonal_decompose
import numpy as np

# Load data
df = pd.read_csv('/content/historical_weather.csv')

# Group the DataFrame by 'city_id'
grouped = df.groupby('city_id')

# Create a dictionary to hold DataFrames for each city
city_datasets = {city_id: group_df for city_id, group_df in grouped}

# Define a function to fill NaN values using seasonal decomposition
def fill_na_with_seasonal_decomposition(series, period=7):
    # Initial fill to handle NaNs for decomposition
    series_filled = series.fillna(method='ffill').fillna(method='bfill')
    decomposition = seasonal_decompose(series_filled, model='additive', period=period)
    filled_residuals = decomposition.resid.interpolate(method='linear')
    return (decomposition.trend + decomposition.seasonal + filled_residuals).fillna(method='bfill').fillna(method='ffill')

# Prepare the submission DataFrame
submission_df = pd.DataFrame(columns=['submission_ID', 'city_id', 'date', 'avg_temp_c'])
submission_id = 1

for city_id, df in city_datasets.items():
    print(f"Processing city: {city_id}")

    # Convert date column to datetime
    df['date'] = pd.to_datetime(df['date'])
    # Set date as index
    df.set_index('date', inplace=True)

    # Apply the function to each relevant column
    df['avg_temp_c'] = fill_na_with_seasonal_decomposition(df['avg_temp_c'])
    df['avg_wind_dir_deg'] = fill_na_with_seasonal_decomposition(df['avg_wind_dir_deg'])
    df['avg_wind_speed_kmh'] = fill_na_with_seasonal_decomposition(df['avg_wind_speed_kmh'])

    # Transform wind direction to sine and cosine components
    df['wind_dir_sin'] = np.sin(np.deg2rad(df['avg_wind_dir_deg']))
    df['wind_dir_cos'] = np.cos(np.deg2rad(df['avg_wind_dir_deg']))

    # Ensure no NaN values remain
    df.fillna(0, inplace=True)

    # Reset the index to get back the date column
    df.reset_index(inplace=True)

    # Columns to forecast
    columns_to_forecast = ['avg_temp_c']

    # Initialize the new DataFrame for predictions
    future_dates = pd.date_range(start='2019-01-01', periods=7, freq='D')
    new_df = pd.DataFrame({'date': future_dates, 'city_id': city_id})

    # Train and predict for each column
    for column in columns_to_forecast:
        print(f"Training model for {column} for city {city_id}")
        temp_df = df[['date', column]].rename(columns={'date': 'ds', column: 'y'})

        # Train the model
        m = Prophet()
        m.fit(temp_df)

        # Make future predictions
        future = m.make_future_dataframe(periods=7)
        forecast = m.predict(future)

        # Extract the predictions for the next 7 days
        forecast = forecast[['ds', 'yhat']].tail(7).reset_index(drop=True)
        new_df[column] = forecast['yhat']

    # Combine the original and new DataFrames
    df['city_id'] = city_id
    df = df.rename(columns={'date': 'ds'})

    # Reorder columns to match the new DataFrame format
    df = df[['ds', 'city_id', 'avg_temp_c', 'avg_wind_speed_kmh', 'wind_dir_sin', 'wind_dir_cos']]

    # Concatenate the original and new DataFrames
    final_df = pd.concat([df, new_df.rename(columns={'date': 'ds'})], ignore_index=True)

    # Ensure no NaN values remain
    final_df.fillna(0, inplace=True)
    # Reset the index to get back the date column
    final_df.reset_index(inplace=True)

    final_df = final_df[['ds', 'avg_temp_c', 'avg_wind_speed_kmh', 'wind_dir_sin', 'wind_dir_cos']]
    final_df.columns = ['ds', 'y', 'x1', 'x2', 'x3']

    train = final_df.iloc[:len(final_df) - 7]
    test = final_df.iloc[len(final_df) - 7:]

    m = Prophet()

    m.add_regressor('x1')
    m.add_regressor('x2')
    m.add_regressor('x3')

    m.fit(train)
    future = m.make_future_dataframe(periods=7)
    future['x1'] = final_df['x1']
    future['x2'] = final_df['x2']
    future['x3'] = final_df['x3']

    forecast = m.predict(future)
    forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(7)

    # Add predicted values to submission DataFrame
    submission_data = {
        'submission_ID': range(submission_id, submission_id + 7),
        'city_id': city_id,
        'date': future_dates,
        'avg_temp_c': new_df['avg_temp_c']
    }
    submission_id += 7
    submission_df = pd.concat([submission_df, pd.DataFrame(submission_data)], ignore_index=True)

# Save to CSV
submission_df.to_csv('submission2.csv', index=False)

print("Forecasting and submission file generation complete.")
```

```python
import pandas as pd
import numpy as np
from prophet import Prophet
from statsmodels.tsa.statespace.sarimax import SARIMAX
from xgboost import XGBRegressor
from statsmodels.tsa.seasonal import seasonal_decompose
from joblib import Parallel, delayed
import multiprocessing

# Load data
df = pd.read_csv('/content/historical_weather.csv')

# Fill NaN values using seasonal decomposition
def fill_na_with_seasonal_decomposition(series, period=7):
    series_filled = series.fillna(method='ffill').fillna(method='bfill')
    decomposition = seasonal_decompose(series_filled, model='additive', period=period)
    filled_residuals = decomposition.resid.interpolate(method='linear')
    return (decomposition.trend + decomposition.seasonal + filled_residuals).fillna(method='bfill').fillna(method='ffill')

# Process each city's data
def process_city(city_id, city_df):
    print(f"Processing city: {city_id}")

    city_df['date'] = pd.to_datetime(city_df['date'])
    city_df.set_index('date', inplace=True)

    # Fill missing values for all relevant columns
    for column in ['avg_temp_c', 'min_temp_c', 'max_temp_c', 'avg_wind_dir_deg', 'avg_wind_speed_kmh', 'precipitation_mm']:
        city_df[column] = fill_na_with_seasonal_decomposition(city_df[column])

    # Special handling for snow_depth_mm
    if city_df['snow_depth_mm'].isna().sum() < len(city_df):
        city_df['snow_depth_mm'] = fill_na_with_seasonal_decomposition(city_df['snow_depth_mm'])
    else:
        city_df['snow_depth_mm'].fillna(0, inplace=True)

    # Ensure no NaN values remain
    city_df.fillna(0, inplace=True)

    # Transform wind direction to sine and cosine components
    city_df['wind_dir_sin'] = np.sin(np.deg2rad(city_df['avg_wind_dir_deg']))
    city_df['wind_dir_cos'] = np.cos(np.deg2rad(city_df['avg_wind_dir_deg']))

    city_df.reset_index(inplace=True)

    # Prepare DataFrame for future dates
    future_dates = pd.date_range(start='2019-01-01', periods=7, freq='D')
    new_df = pd.DataFrame({'date': future_dates, 'city_id': city_id})

    # Forecast each column using Prophet, ARIMA, and XGBoost, then ensemble the results
    for column in ['avg_temp_c', 'min_temp_c', 'max_temp_c', 'precipitation_mm', 'snow_depth_mm', 'avg_wind_speed_kmh']:
        print(f"Training models for {column} for city {city_id}")
        temp_df = city_df[['date', column]].rename(columns={'date': 'ds', column: 'y'}).dropna()

        # Prophet Model
        m = Prophet()
        m.fit(temp_df)
        future = m.make_future_dataframe(periods=7)
        prophet_forecast = m.predict(future)['yhat'].tail(7).values

        # ARIMA Model
        temp_df.set_index('ds', inplace=True)
        arima_model = SARIMAX(temp_df['y'], order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
        arima_fit = arima_model.fit(disp=False)
        arima_forecast = arima_fit.forecast(steps=7).values

        # XGBoost Model
        temp_df['month'] = temp_df.index.month
        temp_df['day'] = temp_df.index.day
        xgboost_model = XGBRegressor(tree_method='gpu_hist', predictor='gpu_predictor')
        X_train = temp_df[['month', 'day']]
        y_train = temp_df['y']
        xgboost_model.fit(X_train, y_train)
        X_test = pd.DataFrame({'month': future_dates.month, 'day': future_dates.day})
        xgboost_forecast = xgboost_model.predict(X_test)

        # Ensemble Forecast
        ensemble_forecast = (prophet_forecast + arima_forecast + xgboost_forecast) / 3
        new_df[column] = ensemble_forecast

    return new_df

# Prepare submission DataFrame
submission_df = pd.DataFrame(columns=['submission_ID', 'city_id', 'date', 'avg_temp_c'])
submission_id = 1

# Get the number of CPU cores for parallel processing
num_cores = multiprocessing.cpu_count()

# Process each city's dataset in parallel
results = Parallel(n_jobs=num_cores)(delayed(process_city)(city_id, city_df) for city_id, city_df in df.groupby('city_id'))

# Combine results and prepare the submission DataFrame
for city_forecast in results:
    city_forecast['submission_ID'] = range(submission_id, submission_id + 7)
    submission_id += 7
    submission_df = pd.concat([submission_df, city_forecast[['submission_ID', 'city_id', 'date', 'avg_temp_c']]], ignore_index=True)

# Save submission file
submission_df.to_csv('submission_ensemble.csv', index=False)
print("Forecasting and submission file generation complete.")
```
